```python
#classe principale 
class Snake_ia :
    def __init__(self, head_position, food_position, direction, game_size, snake_positions, reverse_commands, ip, commands_history):
        self.head_x_pos = head_position[0]
        self.head_y_pos = head_position[1]
        self.food_x_pos = food_position[0]
        self.food_y_pos = food_position[1]
        self.direction = direction
        self.game_size = game_size
        self.snake_positions = snake_positions
        self.reverse_commands = reverse_commands
        self.directions_left = []
        self.ip = ip
        self.history = commands_history
        self.is_eaten = False
        
    def remove_wrong_direction(self, x, y):
        #print("direction = " + str(self.direction) +", la position = " + str((x,y)) + str(self.snake_positions[1:]))
        if (x,y) in self.snake_positions[1:]:
            if self.direction in self.directions_left: 
                self.directions_left.remove(self.direction)
       
        
    def shortest_path(self):
        distance_to_food_x = self.head_x_pos - self.food_x_pos
        distance_to_food_y = self.head_y_pos - self.food_y_pos
        
        if distance_to_food_y != 0:
            if distance_to_food_y > 0:
                self.direction = "top"
            else:
                self.direction = "bottom"
        
        if distance_to_food_x != 0:
            if distance_to_food_x > 0:
                self.direction = "left"
            else:
                self.direction =  "right"
        
        return self.direction

    def evolved_shortest_path(self):
        distance_to_food_x = self.head_x_pos - self.food_x_pos
        distance_to_food_y = self.head_y_pos - self.food_y_pos
        directions_left = []

        if distance_to_food_y != 0 :
            if distance_to_food_y > 0: 
                directions_left.append("top")
            else : 
                directions_left.append("bottom")
        else :
            if distance_to_food_x != 0 :
                if distance_to_food_x > 0 :
                    directions_left.append("left")
                else :
                    directions_left.append("right")
  
        if self.direction == "right":
            x = self.head_x_pos + 1
            y = self.head_y_pos
            self.remove_wrong_direction(x, y)
                
        if self.direction == "left":
            x = self.head_x_pos - 1
            y = self.head_y_pos
            self.remove_wrong_direction(x, y)
                
        if self.direction == "top":
            x = self.head_x_pos - 1
            y = self.head_y_pos
            self.remove_wrong_direction(x, y)
                
        if self.direction == "bottom":
            x = self.head_x_pos + 1
            y = self.head_y_pos
            self.remove_wrong_direction(x, y)
        
        if self.reverse_commands[self.direction] in directions_left:
            directions_left.remove(self.reverse_commands[self.direction])
              
        if len(directions_left) == 0:
            directions_left = ["right", "left", "top", "bottom"]
            directions_left.remove(self.reverse_commands[self.direction])
            directions_left.remove(self.direction)

        #print("FINAL", directions_left)    
        return directions_left[-1]
            
    def hamiltonian(self):
        if self.head_y_pos == self.game_size :
            if self.head_x_pos - 1 == 0:
                self.direction = "top"
            else:
                self.direction = "left"
        elif self.direction == "right":
            if self.head_x_pos != self.game_size:
                self.direction = "right"
                pass
            else:
                self.direction = "bottom"
        elif self.direction == "bottom":
            if self.head_x_pos == self.game_size:
                self.direction = "left"
                pass
            else:
                self.direction = "right"
        elif self.direction == "left":
            if self.head_x_pos - 1 != 1:
                self.direction = "left"
            else:
                self.direction = "bottom"
        elif self.direction == "top" and self.head_y_pos - 1 == 0:
            self.direction = "right"

        return self.direction
   

```
